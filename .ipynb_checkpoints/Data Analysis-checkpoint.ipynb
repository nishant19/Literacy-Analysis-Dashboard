# ******WEEK 2******

**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

**Importing Dataset**

```python
elem_1 = pd.read_excel('SRC_Rawdata_2015-16_Elementary.xlsx', sheet_name='2015-16_1', header=19)
elem_1.set_index('STATCD', inplace=True)
```

**Column Selected**

```python
col = ['STATNAME', 'TOTPOPULAT', 'P_URB_POP', 'OVERALL_LI', 'AREA_SQKM',
       'TOT_6_10_15', 'TOT_11_13_15', # projected population
       'SCHTOT', # total schools
       'SCHTOTG',  # total govt schools
       'SCHTOTGR', # total govt schools - rural
       'SCHTOTPR', # total private schools - rural
       'ENRTOT', # elementary enrollment in schools
       'ENRTOTG', # elementary enrollment in schools - govt
       'ENRTOTGR', # elementary enrollment in schools - govt, rural
       'ENRTOTPR', # elementary enrollment in schools - private, rural
       'TCHTOTG', # teachers in schools - govt
       'TCHTOTP', # teachers in schools - private
       'SCLSTOT', # single classroom schools
       'STCHTOT', # single teacher schools
       'ROADTOT', # Schools Approachable by All Weather Road
       'SPLAYTOT', # Schools with Playground Facility
       'SWATTOT', # Schools with Drinking Water
       'SELETOT', # Schools with Electricity
    ] 
```

**Shape of current selected dataset**

```python
elem_1 = elem_1.loc[:,col]
elem_1.shape
```

**Second Dataset**

```python
elem_2 = pd.read_excel('SRC_Rawdata_2015-16_Elementary.xlsx', sheet_name='2015-16_2', header=20)
elem_2.set_index('STATCD', inplace=True)

#Student data
col = [ 'C1_B', 'C2_B', 'C3_B', 'C4_B', 'C5_B', 'C6_B', 'C7_B', 'C8_B', 'C9_B',
       'C1_G', 'C2_G', 'C3_G', 'C4_G', 'C5_G', 'C6_G', 'C7_G', 'C8_G', 'C9_G', # Enrollment by grade(1 to 9) Boys & Girls
       'C1_BD', 'C2_BD', 'C3_BD', 'C4_BD', 'C5_BD', 'C6_BD', 'C7_BD', 'C8_BD',
       'C1_GD', 'C2_GD', 'C3_GD', 'C4_GD', 'C5_GD', 'C6_GD', 'C7_GD', 'C8_GD', # Children with Special Needs grade(1 to 8) Boys & Girls
       'C1_BR', 'C2_BR', 'C3_BR', 'C4_BR', 'C5_BR', 'C6_BR', 'C7_BR', 'C8_BR', 'C9_BR',
       'C1_GR', 'C2_GR', 'C3_GR', 'C4_GR', 'C5_GR', 'C6_GR', 'C7_GR', 'C8_GR', 'C9_GR', # Repeaters By Grade
    ]
```

**Shape of second Data**

```python
elem_2 = elem_2.loc[:,col]
elem_2.shape
```

**Joining both dataset**

```python
elem = elem_1.join(elem_2)
elem.shape
```

**Elementary data processing**

```python
#population real value
elem.TOTPOPULAT = elem.TOTPOPULAT*1000

#population density(total population per unit area)
elem['POP_DENSITY'] = elem.TOTPOPULAT / elem.AREA_SQKM
#total elementary population
elem['TOT_ELEM_POP'] = 1000*(elem.TOT_6_10_15 + elem.TOT_11_13_15) # Ages 6 to 13

#deleting used value
del elem['TOT_6_10_15'] 
del elem['TOT_11_13_15']

#total boys enrollment
col = ['C1_B', 'C2_B', 'C3_B', 'C4_B', 'C5_B', 'C6_B', 'C7_B', 'C8_B', 'C9_B']
elem['TOT_ENROLL_B'] = elem.loc[:, col].sum(axis=1)

#total girls enrollment
col = ['C1_G', 'C2_G', 'C3_G', 'C4_G', 'C5_G', 'C6_G', 'C7_G', 'C8_G', 'C9_G',]
elem['TOT_ENROLL_G'] = elem.loc[:, col].sum(axis=1)

#total specially needed children enrollment
col = ['C1_GD', 'C2_GD', 'C3_GD', 'C4_GD', 'C5_GD', 'C6_GD', 'C7_GD', 'C8_GD']
elem['TOT_SPECIAL_NEEDS'] = elem.loc[:, col].sum(axis=1)

#repeaters in boys
col = ['C1_BR', 'C2_BR', 'C3_BR', 'C4_BR', 'C5_BR', 'C6_BR', 'C7_BR', 'C8_BR', 'C9_BR',]
elem['TOT_REPEATERS_B'] = elem.loc[:, col].sum(axis=1)

#repeaters in girls
col = ['C1_GR', 'C2_GR', 'C3_GR', 'C4_GR', 'C5_GR', 'C6_GR', 'C7_GR', 'C8_GR', 'C9_GR',]
elem['TOT_REPEATERS_G'] = elem.loc[:, col].sum(axis=1)
```

**Dropping used data**

```python
col = [ 'C1_B', 'C2_B', 'C3_B', 'C4_B', 'C5_B', 'C6_B', 'C7_B', 'C8_B', 'C9_B',
       'C1_G', 'C2_G', 'C3_G', 'C4_G', 'C5_G', 'C6_G', 'C7_G', 'C8_G', 'C9_G', # Enrollment by grade(1 to 9) Boys & Girls
       'C1_BD', 'C2_BD', 'C3_BD', 'C4_BD', 'C5_BD', 'C6_BD', 'C7_BD', 'C8_BD',
       'C1_GD', 'C2_GD', 'C3_GD', 'C4_GD', 'C5_GD', 'C6_GD', 'C7_GD', 'C8_GD', # Children with Special Needs grade(1 to 8) Boys & Girls
       'C1_BR', 'C2_BR', 'C3_BR', 'C4_BR', 'C5_BR', 'C6_BR', 'C7_BR', 'C8_BR', 'C9_BR',
       'C1_GR', 'C2_GR', 'C3_GR', 'C4_GR', 'C5_GR', 'C6_GR', 'C7_GR', 'C8_GR', 'C9_GR', # Repeaters By Grade
    ]

elem = elem.drop(axis=0, columns=col)
elem.set_index('STATNAME', inplace=True)


elem.loc['WEST BENGAL', 'TOTPOPULAT'] = elem.loc['WEST BENGAL', 'TOTPOPULAT']/10
elem.loc['WEST BENGAL', 'POP_DENSITY'] = elem.loc['WEST BENGAL', 'TOTPOPULAT'] / elem.loc['WEST BENGAL', 'AREA_SQKM']

# Removing TELANGANA
elem = elem.head(35)
```
